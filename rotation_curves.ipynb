# Rotation Curves

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_ngc7753 = "/home/angeldavid/Desktop/Fedora/RESEARCH-INTERSHIPS/data_cubes/NGC7753/output/SoFiA_J234703.61+292835.6/rings_final2.txt"
file_ngc7752 = "/home/angeldavid/Desktop/Fedora/RESEARCH-INTERSHIPS/data_cubes/NGC7752/output/SoFiA_J234703.61+292835.6/rings_final2.txt"
arp68_3_file = "/home/angeldavid/Desktop/Fedora/RESEARCH-INTERSHIPS/data_cubes/USNOA2-1125-19940303/output0/SoFiA_J234703.61+292835.6/rings_final2.txt"
```

```python
rot_curve_data_1=pd.read_csv(file_ngc7753,delim_whitespace=True,on_bad_lines='skip')
rot_curve_data_2=pd.read_csv(file_ngc7752,delim_whitespace=True,on_bad_lines='skip')
rot_curve_data_3=pd.read_csv(arp68_3_file,delim_whitespace=True,on_bad_lines='skip')
```

```python
radius_arcsec1 = rot_curve_data_1["RAD(arcs)"]
rot_vel1 = rot_curve_data_1['VROT(km/s)']
radius_arcsec1 = radius_arcsec1.astype(float)
rot_vel1 = rot_vel1.astype(float)
radius_arcsec1 = np.insert(radius_arcsec1.values, 0, 0)
rot_vel1 = np.insert(rot_vel1.values, 0, 0)
print(radius_arcsec1)
#print(rot_vel)
```

```python
radius_arcsec2 = rot_curve_data_2["RAD(arcs)"]
rot_vel2 = rot_curve_data_2['VROT(km/s)']
radius_arcsec2 = radius_arcsec2.astype(float)
rot_vel2 = rot_vel2.astype(float)
radius_arcsec2 = np.insert(radius_arcsec2.values, 0, 0)
rot_vel2 = np.insert(rot_vel2.values, 0, 0)
print(radius_arcsec2)
```

```python
radius_arcsec3 = rot_curve_data_3["RAD(arcs)"]
rot_vel3 = rot_curve_data_3['VROT(km/s)']
radius_arcsec3 = radius_arcsec3.astype(float)
rot_vel3 = rot_vel3.astype(float)
radius_arcsec3 = np.insert(radius_arcsec3.values, 0, 0)
rot_vel3 = np.insert(rot_vel3.values, 0, 0)
print(radius_arcsec3)
```

```python
# Optional: set a consistent font and style
plt.rcParams.update({
    'font.size': 14,
    'font.family': 'serif',
    'axes.linewidth': 1.2,
    'xtick.major.size': 6,
    'ytick.major.size': 6,
    'xtick.direction': 'in',
    'ytick.direction': 'in'
})

# Create figure
fig, ax = plt.subplots(figsize=(8, 6))

# Plot rotation curves
ax.plot(radius_arcsec1, rot_vel1, linestyle='-', color='blue', label='NGC7753', linewidth=2, markersize=5)
ax.plot(radius_arcsec2, rot_vel2, linestyle='-', color='red' , label='NGC7752', linewidth=2, markersize=5)
ax.plot(radius_arcsec3, rot_vel3, linestyle='-', color='green', label='USNOA2-1125', linewidth=2, markersize=5)

# Labels and title
ax.set_xlabel('Radius (arcsec)', fontsize=16)
ax.set_ylabel('Rotation Velocity (km/s)', fontsize=16)
ax.set_title('Rotation Curves', fontsize=18)

# Limits and ticks
ax.set_xlim(0, 100)
ax.set_ylim(0, 300)
ax.set_xticks(range(0, 110, 10))
ax.set_yticks(range(0, 350, 50))

# Grid and legend
#ax.grid(True, linestyle='--', linewidth=0.6, alpha=0.7)
ax.legend(fontsize=12, loc='upper right', frameon=False)

# Improve layout
plt.tight_layout()
# Optional: save as high-resolution figure
plt.savefig('rotation_curves.jpg', dpi=300)
plt.show()
```
